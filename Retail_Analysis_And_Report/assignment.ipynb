```python
cols_prv=pricing_data.columns
cols=[x.lower().replace('_%','').replace('%','').replace(' ','_') for x in pricing_data.columns]
# pricing_data.rename(columns=)
cols_dict=dict(zip(cols_prv,cols))
cols_dict
```

```python
pricing_data.rename(columns=cols_dict,inplace=True)
pricing_data.head()
```

```python
pricing_data[pricing_data['handling_cost'].isna()]['handling_cost']='$0.75'
```

```python

currency_cols = ['fba_fee', 'storage_fee', 'handling_cost', 'cost','current_price']
percent_cols = ['minimum_acceptable_margin', 'target_gross_margin']

pricing_data = (
    pricing_data
    .assign(
        **{
            col: pricing_data[col]
                    .astype(str)
                    .str.replace(r'[\$,]', '', regex=True)
                    .astype(float)
            for col in currency_cols
        },
        **{
            col: pricing_data[col]
                    .astype(str)
                    .str.replace('%', '', regex=True)
                    .astype(float)
                    .div(100)
            for col in percent_cols
        }
    )
)

```
